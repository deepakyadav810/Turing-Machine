```python
N = 1000 # tape length, initialize to a large value
 
class TuringMachine:
 
    '''
    initialize the Turing Machine, read the program and input
    '''
    def __init__(self, program, input, state=0):
        self.trf = {}
        self.state = str(state)#initail state is 0
        self.tape = ''.join(['_']*N)#creating tape of 1000length
        self.head = N // 2   # head is positioned in the middle
        self.tape = self.tape[:self.head] + input + self.tape[self.head:] #adding the input to the tape
        for line in program.splitlines(): #storing all the transition info in dictonary
            s, a, r, d, s1 = line.split(' ')
            self.trf[s,a] = (r, d, s1)
 
    '''
    step through a program
    '''
    def step(self):
        if self.state != 'H':
            # assert self.head >= 0 and self.head < len(self.tape) here
            """
            print("Head",self.head)
            print("Tape",self.tape)
            print("TRF",self.trf)
            print("before+",self.tape[:self.head])
            print("after+",self.tape[self.head:])
            """
            a = self.tape[self.head] #value of current head position in tape
            #print("a",a)
            action = self.trf.get((self.state, a)) #matching the dictonary to perform action
            #print("action",action)
            if action:
                r, d, s1 = action
                self.tape = self.tape[:self.head] + r + self.tape[self.head+1:]##now adding the new element in place of previous element by just shifting the head position by 1
                #print("New Tape",self.tape)
                if d != '*':
                    self.head = self.head + (1 if d == 'r' else -1)
                self.state = s1
                print(self.tape.replace('_', ''), self.state)
 
    '''
    run a program
    '''
    def run(self, max_iter=9999):
        iter = 0
        while self.state != 'H' and iter < max_iter: # prevent infinite loop
            self.step()
            iter += 1
```

```python
input = '1101_101'
program = open('transition1.txt').read()
tm = TuringMachine(program, input)
tm.run()
# 10010 H
```
